# 데이터 시각화

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.rcParams["font.family"] = 'Malgun Gothic'
plt.rcParams["font.size"] = 10
```

### 1. 라인 플롯

```python
plt.plot([1, 4, 9, 16])   # x = [0, 1, 2, 3]임을 가정
plt.show()
```

```python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16])    # x축 값, y축 값
plt.title('y = x**2 그래프')
plt.grid()   # 격자
plt.show()
```

- 스타일 지정  : 색깔(color), 마커(marker), 선 종류(line style)의 순서로 지정

```python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16],'rs--')
plt.title("'rs--' 스타일")
plt.grid()
plt.show()
```

```python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16], c='c', lw = 3, marker='o', ms = 10)
plt.title("'rs--' 스타일")
plt.grid()
plt.show()
```

- 그림 범위 지정

```python
plt.plot([10, 20, 30, 40], [1, 4, 9, 16],
         c="b", lw=5, ls="--", marker="o", ms=15, mec="g", mew=5, mfc="r")
plt.title("x축, y축의 범위 설정")
plt.xlim(0, 50)
plt.ylim(-2, 20)
plt.grid()
plt.show()
```

- 틱 설정 : 플롯이나 차트에서 축상의 위치 표시 지점을 틱(tick)이라고 하고 이 틱에 써진 숫자 혹은 글자를 틱 라벨(tick label)이라고 함

```python
import matplotlib as mpl
mpl.rcParams['axes.unicode_minus'] = False 
```

```python
X = np.linspace(-np.pi, np.pi, 300)
Y = np.cos(X)
plt.plot(X, Y)
plt.grid()
plt.show()
```

```python
X = np.linspace(-np.pi, np.pi, 300)
Y = np.cos(X)
plt.plot(X, Y)
plt.xticks([-np.pi, -np.pi / 2, 0, np.pi / 2, np.pi])
plt.yticks([-1, 0, 1])
plt.title("x축, y축의 tick label 설정")
plt.grid()
plt.show()
```

```python
X = np.linspace(-np.pi, np.pi, 300)
Y = np.cos(X)
plt.plot(X, Y)
plt.xticks([-np.pi, -np.pi / 2, 0, np.pi / 2, np.pi],
           [r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi/2$', r'$+\pi$'])   # $$ 사이에 LaTeX 수학 문자식을 넣을 수 있음
plt.yticks([-1, 0, 1], ["Low", "Zero", "High"])
plt.title("x축, y축의 tick label 설정")
plt.grid()
plt.show()
```

- 여러개의 선 그리기

라인 플롯에서 선을 하나가 아니라 여러개를 그리고 싶은 경우에는 x 데이터, y 데이터, 스타일 문자열을 반복하여 인수로 넘김

```python
t = np.arange(0., 5., 0.2)
plt.title("라인 플롯에서 여러개의 선 그리기")
plt.plot(t, t, 'r--', t, 0.5 * t**2, 'bs:', t, 0.2 * t**3, 'g^-')
plt.grid()
plt.show()
```

- 복수 개의 plot을 사용하여 겹쳐 그리기

```python
plt.title("라인 플롯에서 여러개의 선 그리기")
plt.plot(t, t, 'r--')
plt.plot(t, 0.5 * t**2, 'bs:')
plt.plot(t, 0.2 * t**3, 'g^-')
plt.grid()
plt.show()
```

- 범례(Legend)

```python
X = np.linspace(-np.pi, np.pi, 256)
C, S = np.cos(X), np.sin(X)
plt.title("legend를 표시한 플롯")
plt.plot(X, C, ls="--", label="cosine")
plt.plot(X, S, ls=":", label="sine")
plt.legend(loc=2)
plt.grid()
plt.show()
```

- x축, y축 라벨, 타이틀

플롯의 x축 위치와 y축 위치에는 각각 그 데이터가 의미하는 바를 표시하기 위해 라벨(label)를 추가할 수 있음

```python
X = np.linspace(-np.pi, np.pi, 256)
C, S = np.cos(X), np.sin(X)
plt.title("X축, Y축 라벨을 표시한 플롯")
plt.plot(X, C, ls="--", label="cosine")
plt.plot(X, S, ls=":", label="sine")
plt.legend(loc=2)
plt.xlabel('X축'); plt.ylabel('Y축 라벨')
plt.grid()
plt.show()
```
