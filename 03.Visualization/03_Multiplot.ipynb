## 여러개의 그림 그리기

맷플롯리브가 그리는 그림은 Figure 객체, Axes 객체, Axis 객체 등으로 구성. Figure 객체는 한 개 이상의 Axes 객체를 포함하고 Axes 객체는 다시 두 개 이상의 Axis 객체를 포함

```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.rcParams["font.family"] = 'Malgun Gothic'
mpl.rcParams['axes.unicode_minus'] = False 
```

- Figure 객체

```python
np.random.seed(2021)
plt.figure(figsize=(10,2))
plt.title('figure size : (10, 2)')
plt.plot(np.random.randn(100))
plt.grid()
plt.show()
```

- Axes 객체와 subplot

```python
x1 = np.linspace(0.0, 5.0)   # 50개가 default
x2 = np.linspace(0.0, 2.0)
y1 = np.cos(2 * np.pi * x1) * np.exp(-x1)
y2 = np.cos(2 * np.pi * x2)
```

```python
ax1 = plt.subplot(2, 1, 1)
plt.plot(x1, y1, 'yo-')
plt.title('2 subplots')
plt.ylabel('Damped oscillation')
plt.grid()

ax2 = plt.subplot(2, 1, 2)
plt.plot(x2, y2, 'r.-')
plt.ylabel('Undamped')
plt.xlabel('time (sec)')

plt.tight_layout()
plt.grid()
plt.show()
```

```python
ax1 = plt.subplot(211)
plt.title('2 subplots')
ax1.plot(x1, y1, 'yo-')
ax1.set_ylabel('Damped oscillation')

ax2 = plt.subplot(212)
ax2.plot(x2, y2, 'r.-')
ax2.set_ylabel('Undamped')

plt.xlabel('time (sec)')
ax1.grid(); ax2.grid()
plt.show()
```

```python
np.random.seed(0)

plt.subplot(221)
plt.plot(np.random.rand(5))
plt.title("axes 1")

plt.subplot(222)
plt.plot(np.random.rand(5))
plt.title("axes 2")

plt.subplot(223)
plt.plot(np.random.rand(5))
plt.title("axes 3")

plt.subplot(224)
plt.plot(np.random.rand(5))
plt.title("axes 4")

plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(2, 2)

np.random.seed(0)
axes[0, 0].plot(np.random.rand(5))
axes[0, 0].set_title("axes 1")
axes[0, 1].plot(np.random.rand(5))
axes[0, 1].set_title("axes 2")
axes[1, 0].plot(np.random.rand(5))
axes[1, 0].set_title("axes 3")
axes[1, 1].plot(np.random.rand(5))
axes[1, 1].set_title("axes 4")

plt.tight_layout()
plt.show()
```

```python
# Axes 객체를 사용해 반복문으로 그래프 만들기

np.random.seed(0)

fig, axes = plt.subplots(2, 2)
for i in range(4):
    ax = axes[i//2, i%2]   # 행 인덱스 : i//2, 열 인덱스 : i%2
    ax.plot(np.random.rand(5))
    ax.set_title(f"axes {i+1}")

plt.tight_layout()
plt.show()
```

```python
# 경우에 맞춰 사용

np.random.seed(0)

fig, axes = plt.subplots(2, 4)
for i in range(8):
    ax = axes[i//4, i%4]
    ax.plot(np.random.rand(5))
    ax.set_title(f"axes {i+1}")

plt.tight_layout()
plt.show()
```

```python
np.random.seed(0)

fig, axes = plt.subplots(2, 2)
for i in range(4):
    ax = axes[i//2, i%2]   # 행 인덱스 : i//2, 열 인덱스 : i%2
    ax.plot(np.random.rand(5))
    ax.set_title(f"axes {i+1}")
    ax.grid()

plt.tight_layout()
plt.show()
```
