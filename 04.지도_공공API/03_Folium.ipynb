```python
map = folium.Map(location=[37.559868, 126.967109], zoom_start=14)
folium.Marker(
    location=[37.559868, 126.967109], 
    popup='한국경제신문사'
).add_to(map)
folium.Marker(location=[37.566317, 126.977829], 
    tooltip='서울특별시청', 
    icon=folium.Icon(color="red", icon="info-sign")
).add_to(map)
folium.Circle(location=[37.564120, 126.998009], 
    radius=50, tooltip='중구청', 
    color='crimson', 
    fill= False
).add_to(map)
map
```

- addr2.csv를 읽어서 지도 위에 표시하기

```python
df =pd.read_csv('addr2.csv')
df
```

```python
map = folium.Map(location=[df['위도'].mean(), df['경도'].mean()], zoom_start=13)
for i in df.index:
    folium.Marker(
    location=[df['위도'][i], df['경도'][i]], 
    tooltip=df['주소'][i],
    popup=df['건물명'][i]
).add_to(map)
title_html ='<h3 align="center" style="font-size:20px"><b>5개 지역 표시</b></h3>'
map.get_root().html.add_child(folium.Element(title_html))
map
```
