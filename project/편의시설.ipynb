# 행정구 편의시설

```python
import folium
```

```python
import json
geo_str = json.load(open('skorea_municipalities_geo_simple.json',
                         encoding='utf8'))
```

```python
import pandas as pd 
df = pd.read_csv('음식료품소매점_구별.csv', encoding= 'cp949')
df.head()
```

```python
map = folium.Map(
    location=[37.561, 126.986], 
    zoom_start=11, tiles='cartodb positron'
)
folium.Choropleth(
    geo_data=geo_str,
    data=df,
    columns=('행정구','편의시설개수'),
    fill_color='PuRd',
    key_on='feature.properties.name',
    legend_name='편의시설(수)'
).add_to(map)

map
```
