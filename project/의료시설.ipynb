```python
import json
geo_str = json.load(open('skorea_municipalities_geo_simple.json',
                         encoding='utf8'))
```

```python
import pandas as pd 
df = pd.read_csv('의료시설_구.csv', encoding= 'cp949')
df.head()
```

```python
import folium

map = folium.Map(
    location=[37.561, 126.986], 
    zoom_start=11, tiles='cartodb positron'
)

folium.Choropleth(
    geo_data=geo_str,
    data=df,
    columns=('행정구','개수'),
    key_on='feature.properties.name',
    fill_color='PuRd'
).add_to(map)

map
```
