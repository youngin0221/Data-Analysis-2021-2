# 행정구 복지시설

```python
import folium
```

```python
import json
geo_str = json.load(open('skorea_municipalities_geo_simple.json',
                         encoding='utf8'))
```

```python
import pandas as pd 
df = pd.read_csv('노인복지시설_취합.csv', encoding= 'cp949')
df.head()
```

```python
group = df.groupby(["행정구"])
df_count = group.size().reset_index(name='복지시설수')
df_count
```

```python
map = folium.Map(
    location=[37.561, 126.986], 
    zoom_start=11, tiles='cartodb positron'
)
folium.Choropleth(
    geo_data=geo_str,
    data=df_count,
    columns=('행정구','복지시설수'),
    fill_color='PuRd',
    key_on='feature.properties.name',
    legend_name='복지시설(수)'
).add_to(map)

map
```
