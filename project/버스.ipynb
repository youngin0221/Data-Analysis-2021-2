```python
import folium
import pandas as pd
```

## 구별

```python
import json
geo_str = json.load(open('skorea_municipalities_geo_simple.json',
                         encoding='utf8'))
```

```python
df = pd.read_csv('버스정류소_구.csv', encoding= 'cp949')
df.head()
```

```python
map = folium.Map(
    location=[37.561, 126.986], 
    tiles='cartodb positron', zoom_start=11
)
folium.Choropleth(
    geo_data=geo_str,
    data=df,
    columns=('행정구','개수'),
    key_on='feature.properties.name',
    fill_color='PuRd'
).add_to(map)
map
```
